```python
from pydub import AudioSegment
import pydub.scipy_effects
import pydub
import io
import os
```

```python
directory = './large_files/audio'
root = 'C:/Users/Blaine/Python/Project_3_ML/'
export = 'C:/Users/Blaine/Python/Project_3_ML/large_files'
```

```python
for filename in os.listdir(directory):
        os.chdir(root)
        file = os.path.join(directory, filename)
        folder = filename
        os.mkdir('./large_files/filtered_audio/' + filename)
        for filename in os.listdir(file):
            os.chdir(root)
            print(file + '/' + folder + '/' + filename)
            path = file[0:19] + '/' + folder + '/' + filename
            song = AudioSegment.from_wav(path)
            high_pass = AudioSegment.high_pass_filter(song, 80, order=2)
            low_pass = AudioSegment.low_pass_filter(high_pass, 300, order=2)
            os.chdir(export)
            low_pass.export("./filtered_audio/" + folder + '/' + filename, format="wav")
```
